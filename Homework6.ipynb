<font size = 3> <b> Упр.3 </b> </font>
Нарисуй графики процентного изменения цены данного эмитента с начала периода. С легендой.

```python
import csv
import matplotlib.pyplot as plt
```

```python
data = []
with open( './files/MTLR_180101_190110.txt' ) as f:
    data_rows = csv.reader( f, delimiter = ';' )
    data = list( data_rows )
close = list( map( lambda x: float(x[-2]), data) )
```

```python
#Преобразуем в проценты
close=list( map(lambda x:100*(x/close[0]), close) )
```

```python
plt.plot( close ,label="Процентное изменение цены с начала периода") 
plt.ylabel( 'Цена в процентах') 
plt.xlabel( 'День начиная с 2018 г') 
plt.legend();
```

<font size = 3> <b> Упр.5 </b> </font>
Написать ламбда функцию которая вычисляет значение квадратичной функции по заданным коэффициентам и точке. Потом применить частичную задачу аргументов для коэффициентов.

```python
import functools as ft
```

```python
#коэффициенты
a=1
b=2
c=1
```

```python
fquad = lambda a, b, c, x: a*x**2+b*x+c
```

```python
#Применяем partial фиксируя коффициенты
fquad_x=ft.partial(fquad,a,b,c)
```

```python
fquad_x(1), fquad(a,b,c,1)
```

```python
x_l=list(range(-10,10))
```

```python
plt.plot( x_l, list(map(fquad_x, x_l)),label= "f(x) = {}x\u00B2+{}x+{}".format(a,b,c)); 
plt.xlabel('Точки');
plt.ylabel('Значения');
plt.legend();
```
