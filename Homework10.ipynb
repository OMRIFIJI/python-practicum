Вычислить гистограмму совсем вручную: использую цикл и список...

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = np.random.randint(-10, 10, 100)
data
```

```python
edges = list(range(-10, 11, 1))
edges
```

```python
hhh = plt.hist( data, edges )
plt.xticks( edges );
```

```python
func = lambda a,b,c: a<=b<c
```

```python
amount=[0]*len(edges)
for d_el in data:
    for ed_el in edges:
        if d_el == ed_el:
            amount[10+ed_el]+=1
            break
```

```python
hhh = plt.hist( data, edges, label="С помощь plt.hist" ) 
hh = plt.bar( edges,amount, width = 0.5, color='black', align = 'edge', label="Построена вручную" )
plt.legend()
plt.xticks( edges );
```

<font size = 3> <b> Упр.2 </b> </font> 
Как циклом за один проход найти наиболее часто встречаемый элемент?

```python
a=np.array([2,3,4,5,6,1,7,8,9,1,0])
```

```python
max_n=0
ans=0
for el in a:
    #Количество совпадений с el
    if len( np.where(a==el)[0] )>max_n:
        ans=el
        max_n=len( np.where(a==el) )
```
